Checking eithere there are missing value present or not in the dataset

```python
df.isnull().sum()
```

```python
#In the column named by "company" there are missing value more than 90% of the total data so we remove this column
```

```python
df.drop(columns="company",inplace=True)
```

```python
#In the column named by agent there are more than 16k value are missing.So, we have to study this column before replacing or removing the missing values.

#Bassically this "agent" column represent the ID of the agent through which the hotel was booked by the customers. 
# So, we cann't replace the ID of agent randomly(by central tendency's value) as it affect our overall analysis because if we replace it by meidan or mode that particular agent may get benefit in this analysis.
#So, we have to analysis the data by removing the missing vlaue and by removing the entire columns so that other data doesnot get affected. 
```

```python
df1=df #copied the dataset
df1
```

```python
df1 = df1.dropna(subset=["agent"])
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
#column named by children has only two missing value so we replace it by meidan
```

```python
median=df1["children"].median()
df1["children"]=df1["children"].fillna(median)
```

```python
#There are 154 null values in the column named by "country", it not less than 1% of the data so we remove those missing values
#so we we be replacing it with mode as the country is the categorical variable
```

```python
df1.dropna(inplace=True)
```

```python
df1.isnull().sum()
```
